```python
import pandas as pd
from pathlib import Path

data_schools_path = Path("Resources/schools_complete.csv")
data_student_path = Path("Resources/students_complete.csv")

school_data = pd.read_csv(data_schools_path)
student_data = pd.read_csv(data_student_path)
school_data
```

```python

student_data
```

```python
student_data.columns
```

```python
data_complete = pd.merge(school_data,student_data,how="left",on=["school_name","school_name"])
data_complete.head()
```

## District Summary

```python
#School count
total_school = len(pd.unique(data_complete["school_name"]))
total_school
#Student count
total_students = len(pd.unique(data_complete["Student ID"]))
total_students
#total budget
total_budget = school_data["budget"].sum()
total_budget
#Average Math score
avg_math = data_complete["math_score"].mean()
avg_math
#Average Reading Score
avg_read = data_complete["reading_score"].mean()
avg_read
#% passing math(% of who passed)
math_count = data_complete[(data_complete["math_score"] >= 70)].count()["student_name"]
math_count_percentage = math_count / float(total_students)*100
math_count_percentage
#% passing reading (% of who passed)
read_count = data_complete[(data_complete["reading_score"] >= 70)].count()["student_name"]
read_count_percentage = read_count/float(total_students) * 100
read_count_percentage
#% overall passing (% of students who passed math AND reading)
mathread_count = data_complete[
    (data_complete["math_score"] >= 70) & (data_complete["reading_score"] >= 70)].count()["student_name"]
mathread_count_percentage = mathread_count/float(total_students) * 100
mathread_count_percentage
```

```python
data =pd.DataFrame({
    'Total Schools' : [total_school],
    'Total Students' : [total_students],
    'Total Budget' : [total_budget],
    'Avg Math Score' : [avg_math],
    'Avg Reading Score' : [avg_read],
    'Percentage of Passing (Math)' : [math_count_percentage],
    'Percentage of Passing (Reading)' : [read_count_percentage],
    'Overall Passing Percentage' : [mathread_count_percentage],    
})
# df = pd.DataFrame(data)

# Formatting
data["Total Students"] = data["Total Students"].map("{:,}".format)
data["Total Budget"] = data["Total Budget"].map("${:,.2f}".format)
data
```

## School Summary

```python
# Use the code provided to select the type per school from school_data
school_types = school_data.set_index(["school_name"])["type"]
school_type = school_types.reset_index().rename(columns={0:"School Type"})
school_type
```

```python
# Calculate the total student count per school from school_data
per_school_counts = data_complete["school_name"].value_counts()
total_s = per_school_counts.reset_index().rename(columns={"count":"Total Students"})
total_s
```

```python
# Calculate the total school budget and per capita spending per school from school_data
per_school_budget = data_complete.groupby(["school_name"]).sum()["budget"]
total_school_budget = per_school_budget.reset_index().rename(columns={0:"Total Budget"})
per_school_capita =per_school_budget/per_school_counts
per_capita = per_school_capita.reset_index().rename(columns={0:"Per Capita"})
```

```python
per_capita
```

```python
# Calculate the average test scores per school from school_data_complete
per_school_math = data_complete.groupby(["school_name"])["math_score"].mean()
per_school_reading = data_complete.groupby(["school_name"])["reading_score"].mean()
```

```python
average_math_score = per_school_math.reset_index().rename(columns={0:"Average Math Score"})
average_math_score
```

```python
average_reading_score = per_school_reading.reset_index().rename(columns={0:"Average Reading Score"})
average_reading_score
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_data_complete
students_passing_math = data_complete[(data_complete)["math_score"] >= 70]
students_passing_reading = data_complete[(data_complete)["reading_score"] >= 70]

students_school_passing_math = students_passing_math.groupby(["school_name"]).count()["student_name"]
students_school_passing_reading = students_passing_reading.groupby(["school_name"]).count()["student_name"]

percentage_math = (students_school_passing_math / per_school_counts) * 100
percentage_reading = (students_school_passing_reading / per_school_counts) * 100
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = data_complete[
    (data_complete["reading_score"] >= 70) & (data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
percentage_overall = school_students_passing_math_and_reading/per_school_counts *100
percentage_overall
```

```python
percent_overall = percentage_overall.reset_index().rename(columns={0:"% Overall Passing"})
percent_overall
```

## Scores by School Spening

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita,spending_bins,labels=labels,include_lowest=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["%Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score" : [spending_math_scores],
    "Average Reading Score" : [spending_reading_scores],
    "% Passing Math" : [spending_passing_math],
    "%Passing Reading" : [spending_passing_reading],
    "% Overall Passing" : [overall_passing_spending]
})

# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"],size_bins,labels=labels)
per_school_summary
```

```python
# Calculate averages for the desired columns.
size_math_scores = per_school_summary.groupby(["school_name"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["school_name"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["school_name"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["school_name"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["school_name"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["school_name"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({
    "Average Math Score" : [size_math_scores],
    "Average Reading Score" : [size_reading_scores],
    "% Passing Math" : [size_passing_math],
    "% Passing Reading" : [size_passing_reading],
    "% Overall Passing" : [size_overall_passing]
})

# Display results
size_summary
```
